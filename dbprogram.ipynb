```python
# Database Program

from tkinter import *
import os
import mysql.connector
import sqlite3

root = Tk()
root.title("My database")
root.config(bg = "cyan")
root.geometry("400x600")

# Database

# Connection to database
conn = sqlite3.connect("usersData.db")
x = conn.cursor()

# Cursor
x = conn.cursor()


# Table
x.execute("""CREATE TABLE IF NOT EXISTS info (
         fname text,
         lname text,
         gender text,
         age integer,
         country text,
         occupation text
         )""")


# Functions
def submitting(): # Submit function adds a record
    conn = sqlite3.connect("usersData.db")
    x = conn.cursor()
    
    # Inserting data into table
    x.execute("INSERT INTO info VALUES (:fname, :lname, :gender, :age, :country, :occupation)", 
             {
                 "fname": fname.get(),
                 "lname": lname.get(),
                 "gender": gender.get(),
                 "age": age.get(),
                 "country": country.get(),
                 "occupation": occupation.get()
             })
    
    conn.commit()
    conn.close()
    
    
    # First makes sure textboxes are clear 
    fname.delete(0,END)
    lname.delete(0,END)
    gender.delete(0,END)
    age.delete(0,END)
    country.delete(0,END)
    occupation.delete(0,END)
    
    
    
def displaying(): # Displays students details
    conn = sqlite3.connect("usersData.db")
    x = conn.cursor()
    
    x.execute("SELECT *, oid FROM info") # Select all and PK of records
    record = x.fetchall() # Fetches all /fetchone /fetchmany
    # print(record)
    
    given = ""
    
    # For-Loop
    for n in record:
        given += str(n[6]) + " " + str(n[0]) + " " + str(n[1]) + "\n"
        
    showLabel = Label(root, text = given)
    showLabel.grid(row = 11, column = 0, columnspan = 2)
    
    conn.commit()
    conn.close()

    
    
def editor(): # Allows user to update a record
    conn = sqlite3.connect("usersData.db")
    x = conn.cursor()
    
    record_id = deleteBox.get()
    
    x.execute("""UPDATE info SET 
        fname = :name,
        lname = :surname,
        gender = :status,
        age = :number,
        country = :place,
        occupation = :work
    
        WHERE oid = :oid""",
              {
                  'name': fnameEdit.get(),  # Dictionary (declares what was shown prev)
                  'surname': lnameEdit.get(),
                  'status': genderEdit.get(),
                  'number': ageEdit.get(),
                  'place': countryEdit.get(),
                  'work': occupationEdit.get(),
                  'oid': record_id
              })
    
    conn.commit() # saves changes
    conn.close() # closes connection
    
    
    # First makes sure textboxes are clear 
    #fnameEdit(0,END)
    #lnameEdit(0,END)
    #genderEdit(0,END)
    #ageEdit(0,END)
    #countryEdit(0,END)
    #occupationEdit(0,END) 
    
    edit.destroy() # After update saved, window closes
    
    
def updating(): # Updates record in a new window
    global edit
    edit = Tk()
    edit.title("Edit databases records")
    edit.config(bg = "cyan")
    edit.geometry("400x200")
    
    conn = sqlite3.connect("usersData.db")
    x = conn.cursor()
    
    
    record_id = deleteBox.get()

    x.execute("SELECT * FROM info WHERE oid = " + record_id)
    record = x.fetchall()
    
    
    # Creates Global variables for textboxes names
    global fnameEdit
    global lnameEdit
    global genderEdit
    global ageEdit
    global countryEdit
    global occupationEdit
    

    # Textboxes
    fnameEdit = Entry(edit, width = 40)
    fnameEdit.grid(row = 0 , column = 1, padx = 20, pady = (10, 0))

    lnameEdit = Entry(edit, width = 30)
    lnameEdit.grid(row = 1 , column = 1, padx = 20)

    genderEdit = Entry(edit, width = 30)
    genderEdit.grid(row = 2 , column = 1, padx = 20)

    ageEdit = Entry(edit, width = 30)
    ageEdit.grid(row = 3 , column = 1, padx = 20)

    countryEdit = Entry(edit, width = 30)
    countryEdit.grid(row = 4 , column = 1, padx = 20)

    occupationEdit = Entry(edit, width = 30)
    occupationEdit.grid(row = 5 , column = 1, padx = 20)


    # Textbox labels
    fnameLabelEdit = Label(edit, text = "First Name: ")
    fnameLabelEdit.grid(row = 0, column = 0, pady = (10, 0))

    lnameLabelEdit = Label(edit, text = "Last Name: ")
    lnameLabelEdit.grid(row = 1, column = 0)

    genderLabelEdit = Label(edit, text = "Gender: ")
    genderLabelEdit.grid(row = 2, column = 0)

    ageLabelEdit = Label(edit, text = "Age: ")
    ageLabelEdit.grid(row = 3, column = 0)

    countryLabelEdit = Label(edit, text = "Country: ")
    countryLabelEdit.grid(row = 4, column = 0)

    occupationLabelEdit = Label(edit, text = "Occupation: ")
    occupationLabelEdit.grid(row = 5, column = 0)

    # Save Button on update page, saves edited records
    saveButtonEdit = Button(edit, bg = "green", fg = "white", text = "Save Record", command = editor)
    saveButtonEdit.grid(row = 9, column = 0, columnspan = 2, pady = 10, ipadx = 137)

    # Gets and displays records in new window via for-loop
    for r in record:
        fnameEdit.insert(0, r[0])
        lnameEdit.insert(0, r[1])
        genderEdit.insert(0, r[2])
        ageEdit.insert(0, r[3])
        countryEdit.insert(0, r[4])
        occupationEdit.insert(0, r[5])
        
    conn.commit()
    conn.close()
    

    
def deleting(): # Deletes record
    conn = sqlite3.connect("usersData.db")
    x = conn.cursor()
    
    x.execute("DELETE FROM info WHERE oid = " + deleteBox.get())
    
    deleteBox.delete(0, END)
    
    conn.commit()
    conn.close()
    
    
    
# Textboxes
fname = Entry(root, width = 30)
fname.grid(row = 0 , column = 1, padx = 20, pady = (10, 0))

lname = Entry(root, width = 30)
lname.grid(row = 1 , column = 1, padx = 20)

gender = Entry(root, width = 30)
gender.grid(row = 2 , column = 1, padx = 20)

age = Entry(root, width = 30)
age.grid(row = 3 , column = 1, padx = 20)

country = Entry(root, width = 30)
country.grid(row = 4 , column = 1, padx = 20)

occupation = Entry(root, width = 30)
occupation.grid(row = 5 , column = 1, padx = 20)


# Insert entry box
deleteBox = Entry(root, width = 30)
deleteBox.grid(row = 7, column = 1, padx = 20, pady = 5)


# Textbox labels
fnameLabel = Label(root, text = "First Name: ")
fnameLabel.grid(row = 0, column = 0, pady = (10, 0))

lnameLabel = Label(root, text = "Last Name: ")
lnameLabel.grid(row = 1, column = 0)

genderLabel = Label(root, text = "Gender: ")
genderLabel.grid(row = 2, column = 0)

ageLabel = Label(root, text = "Age: ")
ageLabel.grid(row = 3, column = 0)

countryLabel = Label(root, text = "Country: ")
countryLabel.grid(row = 4, column = 0)

occupationLabel = Label(root, text = "Occupation: ")
occupationLabel.grid(row = 5, column = 0)

# Delete/Insert Label
deleteLabel = Label(root, text = "Select record ID: ")
deleteLabel.grid(row = 7, column = 0, pady = 5)


# Buttons
# Submit button
submitButton = Button(root, bg = "blue", fg = "white", text = "Add Record", command = submitting)
submitButton.grid(row = 6, column = 0, columnspan = 2, pady = 10, padx = 10, ipadx = 137)

# Show records button
showButton = Button(root, bg = "grey", fg = "white", text = "Display Records", command = displaying)
showButton.grid(row = 10, column = 0, columnspan = 2, pady = 10, padx = 10, ipadx = 137)

# Delete records button
deleteButton = Button(root, bg = "red", fg = "white", text = "Delete Record", command = deleting)
deleteButton.grid(row = 8, column = 0, columnspan = 2, pady = 10, ipadx = 137)

# Update records button
updateButton = Button(root, bg = "green", fg = "white", text = "Update Record", command = updating)
updateButton.grid(row = 9, column = 0, columnspan = 2, pady = 10, ipadx = 137)


# Commit and close connection
conn.commit()
conn.close()

# Displays gui
root.mainloop()
```
